```python
import re
def cnt_rows_with_word(word):
    cnt = 0
    for ind, val in amnez_not_null.iteritems():
        if len(re.findall(word, val.lower())) != 0:
            cnt += 1 
    return cnt
```

```python
data = pd.read_csv('train_data.csv', index_col=0).fillna(0)
data.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2)
```

```python
from catboost import CatBoostClassifier
catboost = CatBoostClassifier(learning_rate=0.03, iterations=1500)
```

```python
catboost.fit(X=train.iloc[:, :-1], y=train.iloc[:, -1], eval_set=[test.iloc[:, :-1], test.iloc[:, -1]], cat_features=[0, 1, 4, 5, 6, 7])
```

```python
f_imp = catboost.get_feature_importance(X=data.iloc[:, :-1], y=data.iloc[:, -1], cat_features=[0, 1, 4, 5, 6, 7])
```

```python
f_imp
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test.iloc[:, -1], catboost.predict_proba(test.iloc[:, :-1])[:, 1], pos_label=1)
metrics.auc(fpr, tpr)
```
