```python
from catboost import CatBoostClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train_data.csv', index_col=0).fillna(0)
data.head()
```

```python
train, test = train_test_split(data, test_size=0.2)
```

```python
catboost = CatBoostClassifier()
```

```python
catboost.fit(X=data.iloc[:, :-1], y=data.iloc[:, -1], cat_features=[0, 1, 4, 5, 6, 7])
```

```python
f_imp = catboost.get_feature_importance(X=data.iloc[:, :-1], y=data.iloc[:, -1], cat_features=[0, 1, 4, 5, 6, 7])
```

```python
f_imp
```

```python
real_test = pd.read_csv('test_data.csv', index_col=0).fillna(0)
real_test.head()
```

```python
ans = catboost.predict_proba(real_test)[:, 1]
ans
```

```python
ans_pandas = pd.DataFrame(data={'ID': real_test.index, 'proba': ans}).reset_index(drop=True)
ans_pandas.head()
```

```python
ans_pandas.to_csv('c_fool.csv', index=False)
```
