```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn import metrics
```

```python
data = pd.read_csv('train-exp.csv', index_col=0).fillna(0)
data.head()
```

```python
data.shape
```

```python
train, test = train_test_split(data, test_size=0.2)
```

```python
catboost = CatBoostClassifier(learning_rate=0.03, iterations=2000)
```

```python
catboost.fit(X=train.iloc[:, list(range(8)) + list(range(9, 25))], y=train.iloc[:, 8], eval_set=[test.iloc[:, list(range(8)) + list(range(9, 25))], test.iloc[:, 8]], cat_features=[0, 1, 4, 5, 6, 7] + list(range(8, 24)))
```

```python
f_imp = catboost.get_feature_importance(X=data.iloc[:, list(range(8)) + list(range(9, 25))], y=data.iloc[:, 8], cat_features=[0, 1, 4, 5, 6, 7] + list(range(8, 24)))
```

```python
f_imp
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test.iloc[:, 8], catboost.predict_proba(test.iloc[:, list(range(8)) + list(range(9, 25))])[:, 1], pos_label=1)
metrics.auc(fpr, tpr)
```

```python
catboost = CatBoostClassifier(learning_rate=0.03, iterations=2000)
```

```python
catboost.fit(X=data.iloc[:, list(range(8)) + list(range(9, 25))], y=data.iloc[:, 8], cat_features=[0, 1, 3, 4, 5, 6, 7] + list(range(8, 24)))
```

```python
real_test = pd.read_csv('test-exp.csv', index_col=0).fillna(0)
real_test.head()
```

```python
ans = catboost.predict_proba(real_test)[:, 1]
ans_pandas = pd.DataFrame(data={'ID': real_test.index, 'proba': ans}).reset_index(drop=True)
ans_pandas.to_csv('nastya_c.csv', index=False)
```
