```python
from catboost import CatBoostClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train_data.csv', index_col=0).fillna(0)
data.head(10)
```

```python
train, test = train_test_split(data, test_size=0.2)
```

```python
catboost = CatBoostClassifier()
```

```python
catboost.fit(X=train.iloc[:, :-1], y=train.iloc[:, -1], eval_set=[test.iloc[:, :-1], test.iloc[:, -1]], cat_features=[0, 1, 4, 5, 6, 7])
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test.iloc[:, -1], catboost.predict_proba(test.iloc[:, :-1])[:, 1], pos_label=1)
metrics.auc(fpr, tpr)
```

```python
additional_triggers = [r'беременн', r'аллерг', r'головная боль', r'консультация повторная', r'сохраня(е|ю)тся']
```

```python
import re
def add_all_cols(col_num, data, words):
    rows = []
    for ind, val in data.iloc[:, col_num].iteritems():
        if ind%1000 == 0:
            print(ind)
        new_row = np.zeros((len(words),))
        for num, word in enumerate(words):
            if val == 0:
                new_row[num] = -1            
            elif len(re.findall(word, val.lower())) != 0:
                new_row[num] = True
            else:
                new_row[num] = False
        rows.append(new_row)
    new_rows = pd.DataFrame(rows, index=data.index)
    return pd.concat([new_rows, data], axis=1)
```

```python
new_train = add_all_cols(6, train, additional_triggers)
new_test = add_all_cols(6, test, additional_triggers)
```

```python
new_train.head(10)
```

```python
catboost = CatBoostClassifier(learning_rate=0.03, iterations=2000)
```

```python
catboost.fit(X=new_train.iloc[:, :-1], y=new_train.iloc[:, -1], eval_set=[new_test.iloc[:, :-1], new_test.iloc[:, -1]], cat_features=[0, 1, 2, 3, 4, 5, 6, 9, 10, 11, 12])
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(new_test.iloc[:, -1], catboost.predict_proba(new_test.iloc[:, :-1])[:, 1], pos_label=1)
metrics.auc(fpr, tpr)
```
